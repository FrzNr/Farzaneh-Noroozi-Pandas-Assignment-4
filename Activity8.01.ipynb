```python
#Import the pandas, numpy and matplotlib package
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load the CSV file as a DataFrame
file_url = 'https://raw.githubusercontent.com/PacktWorkshops/The-Pandas-Workshop/master/Chapter08/Data/PUF2020final_v1coll.csv'
data_frame = pd.read_csv(file_url)
```

```python
#For the sake of simplicity, keep only the following columns: REGION, SQFT, BEDROOMS, PRICE 
columns_to_keep = ['REGION', 'SQFT', 'BEDROOMS', 'PRICE']
data_frame = data_frame[columns_to_keep]
```

```python
#Display the first 10 rows of the DataFrame
data_frame.head(10)
```

```python
#Plot a histogram chart for PRICE to see the distribution of house prices
plt.hist(data_frame['PRICE'], bins=30)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of House Prices')
plt.show()
```

```python
#Plot a histogram chart for SQFT to see the distribution of house areas
plt.hist(data_frame['SQFT'], bins=30)
plt.xlabel('Area (SQFT)')
plt.ylabel('Frequency')
plt.title('Distribution of House Areas')
plt.show()
```

```python
#Plot a scatter plot chart for PRICE and SQFT to understand whether there is any correlation between the two
plt.scatter(data_frame['SQFT'], data_frame['PRICE'])
plt.xlabel('Area (SQFT)')
plt.ylabel('Price')
plt.title('Price vs. Area')
plt.show()
```

```python
#Plot a boxplot chart for BEDROOMS on PRICE to understand the distribution of prices across bedroom sizes
plt.figure(figsize=(10, 6))
plt.boxplot(data_frame.groupby('BEDROOMS')['PRICE'].apply(list), labels=data_frame['BEDROOMS'].unique())
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.title('Price Distribution across Bedroom Sizes')
plt.show()
```

```python
#Plot a scatter chart for BEDROOMS and PRICE to understand whether there is any correlation between the two
plt.figure(figsize=(10, 6))
plt.scatter(data_frame['BEDROOMS'], data_frame['PRICE'])
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.title('Scatter Chart: Bedrooms vs Price')
plt.show()
```

```python
#Plot a boxplot chart for REGION on PRICE to understand the distribution of prices across regions
plt.figure(figsize=(10, 6))
data_frame.boxplot(column='PRICE', by='REGION', vert=False)
plt.xlabel('Region')
plt.ylabel('Price')
plt.title('Boxplot Chart: Price Distribution by Region')
plt.show()
```

```python
#Plot a scatter chart for REGION on PRICE to understand whether there is any correlation between the two
plt.figure(figsize=(10, 6))
plt.scatter(data_frame['REGION'], data_frame['PRICE'])
plt.xlabel('Region')
plt.ylabel('Price')
plt.title('Scatter Chart: Price vs Region')
plt.xticks(rotation=90)
plt.show()
```

```python
#Plot a horizontal bar chart for REGION and PRICE by using the pivot function to see how the average price varies across regions
#Calculate the average price by region using pivot_table
avg_price_by_region = data_frame.pivot_table(values='PRICE', index='REGION', aggfunc='mean')

#Plot the horizontal bar chart
plt.figure(figsize=(10, 6))
avg_price_by_region.plot(kind='barh')
plt.xlabel('Average Price')
plt.ylabel('Region')
plt.title('Average Price by Region')
plt.show()
```

```python
#Plot a line chart for REGION and PRICE by using the pivot function to validate the previous
#Calculate the average price by region using pivot_table
avg_price_by_region = data_frame.pivot_table(values='PRICE', index='REGION', aggfunc='mean')

#Plot the line chart
plt.figure(figsize=(10, 6))
avg_price_by_region.plot(kind='line', marker='o')
plt.xlabel('Region')
plt.ylabel('Average Price')
plt.title('Average Price by Region')
plt.grid(True)
plt.show()
```
